```python
sound = pd.read_csv("Corpus_neu.tsv", sep="\t")
sound.head()
```

```python
sound['sound word density'] = sound['sound word density'].astype('float')
```

```python
# all data
sns.histplot(data = sound,
            x = 'Genre',
            y = 'sound word density'); 
```

```python
decades = sns.boxplot(data = sound,
            y = 'Genre', 
            x = 'sound word density',
            color = 'blue');
plt.savefig("Corpus_short_Genre_swd.png") 
```

```python
# by level
sns.boxplot(data = sound,
            x = 'Decade', 
            y = 'sound word density',
            order=["1800-09", "1810-19", "1820-29", "1830-39", "1840-49", "1850-59", "1860-69", "1870-79", "1880-89", "1890-99", "1900-1909"]
           ).set_title('Boxplots');
#plt.savefig("Corpus_Genre2_swd.png") 
```
